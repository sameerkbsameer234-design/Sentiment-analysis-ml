```python
import pandas as pd
import numpy as np
import nltk
import re

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, Classification_report

nltk.download('punkt')
nltk.download('stopwords')
```

# Preprocessing

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z]', ' ', text)

  tokens = word_tokenize(text)

  tokens = [word for word in tokens if word not in stop_words]
  return" ".join(tokens)

  df['clean_text'] =
  df['text'].apply(clean_text)

  df.head()
```

## Explanation

Preprocessing steps:
1. Lowercasing: Ensures uniformity in text.
2. Removing special characters: Eliminates noise.
3. Tokenization: Splits text into words.
4. Stopword Removal: Removes common words (like 'is', 'the') which do not add meaning.

Impact:
These steps improve model accuracy by reducing noise and focusing on meaningful words.

# Feature Extraction

```python
vectorizer = TfidfVectorizer()

x = vectorizer.fit-transform(df['clean_text'])
y = df['emotion']
```

# Explanation

TF-IDF converts text into numerical features based on word importance.
words that appear frequently in a document but less across all documents get higher scores.This helps models focus on meaningful words rather than common ones.

# Train-Test-Split

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

# Train Naive Bayes

```python
nb_model = MultinomialNB()
nb_model.fit(x_train, y_train)

nb_pred = nb_model.predict(x_test)

print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
print(classification_report(y_test, nb_pred))
```

# Train SVM

```python
svm_model = LinearSVC()
svm_model.fit(x_train, y_train)

svm_pred = svm_model.predict(x_test)

print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
print(classification_report(y_test, svm_pred))
```

# Model Comparison

```python
print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
```

# Final Explanation

Model Comparison:

Naive Bayes:
- Works well for text classification.
- Fast and simple.
- Assumes independence between features.

SVM (Support Vector Machine):
- More powerful for complex data.
- Finds optimal boundary between classes.
- Usually gives higher accuracy than Naive Bayes.

Conclusion:
SVM performed better becuase it handles high-dimensional data effectively, making it suitable for emotion classification tasks.
